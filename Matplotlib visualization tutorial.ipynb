# Bar Plot


```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Nabeel",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=1,label="NABEEL")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Nabeel",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color="r",edgecolor="r",label="Popularity")
plt.bar(x,z,width=0.4,color="b",edgecolor="r",label="Popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("Nabeel",fontsize=15)

plt.bar(p,y,width,color="r",edgecolor="r",label="Popularity")
plt.bar(p1,z,width,color="b",edgecolor="r",label="Popularity1")

plt.xticks(p+width/2,x,rotation=45)
plt.legend()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
plt.scatter(day,No,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
c=["r","b","y","m","y","r","g"]
size=[50,60,200,100,45,20,34]
plt.scatter(day,No,color=c,s=size,marker="*",edgecolor="g")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
color=[10,49,30,29,56,20,30]
size=[50,60,200,100,45,20,34]
plt.scatter(day,No,c=color,s=size,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
No=[2,3,1,4,5,3,6]
color=[10,49,30,29,56,20,30]
size=[50,60,200,100,45,20,34]
plt.scatter(day,No,c=color,s=size,cmap="BrBG")
t=plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

# Pie Chart/Plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","Python"]
plt.pie(x,labels=y)
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","Python"]
ex=[0.3,0.0,0.0,0.0]# makes c appear out
c=["r","b","y","g"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1) 
# radius function changes radius of pie chart
# shadow=true add shadow in graph
# autopct="%0.1f%%"  used for showin percentage in graph if we use 0.2 we will get two decimals example 30.00%
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","java","Python"]
ex=[0.3,0.0,0.0,0.0]# makes c appear out
c=["r","b","y","g"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.3,labeldistance=1.2,startangle=180
       ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":2,"edgecolor":"black"},rotatelabels=False)

plt.legend()
# textprops always should be in dictionary format used to change font size
# start angle changes angle of chart
# radius function changes radius of pie chart
# shadow=true add shadow in graph
# autopct="%0.1f%%"  used for showin percentage in graph if we use 0.2 we will get two decimals example 30.00%
```

# Stack Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,3,2,5,4]

plt.stackplot(x,area)
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"],baseline="zero")# try baseline = wiggle,sym
plt.legend(loc=2)
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

# Step Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,2,3,5,4]

plt.step(x,y,marker="o",color="r",mfc="g",label="Python")#mfc is short form of marker face color

plt.grid()
plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(loc=2)
```

# Fill_between Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x,y)

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2) & (x<=4))

plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
